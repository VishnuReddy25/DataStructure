```python
class circularqueue:
    def __init__(self, maxsize=3):
        self.queue = []
        self.maxsize = maxsize
        self.rear = -1
        self.front = 0
    def enqueue(self, data):
        if self.full():
            return "QF"
        else:
            self.queue.append(data)
            self.rear = (self.rear + 1) % self.maxsize
    def dequeue(self):
        if self.empty():
            return "QE"
        else:
            data = self.queue[self.front]
            self.front = (self.front + 1) % self.maxsize
            return data
    def size(self):
        if self.rear >= self.front:
            size = self.rear - self.front
        else:
            size = self.maxsize - (self.front - self.rear)
        return size
    def full(self):
        return self.size() == self.maxsize

    def empty(self):
        return self.size() == 0
    def peek(self):
        if self.empty():
            return None
        else:
            return self.queue[self.front]
    def display(self):
        if self.rear >= self.front:
            print(self.queue[self.front:])
        else:
            print(self.queue[self.rear:self.front] + self.queue[self.front:])
cq=circularqueue()
cq.enqueue(5)
cq.enqueue(10)
cq.enqueue(50)
print(cq.display())
cq.enqueue(77)
cq.dequeue()
print(cq.display())
cq.enqueue(77)
print(cq.display())
cq.dequeue()
print(cq.display())


```

```python
class doublequeue:
    def __init__(self):
        self.limit=5
        self.q=[None]*5
        self.front=-1
        self.rear=-1

    def isEmpty(self):
        if self.rear==-1:
            return True
        else:
            return False

    def isFull(self):
        if self.rear>=self.limit-1 and self.front==0:
            return True
        else:
            return False

    def enqueue(self,data):
        if self.isFull():
            print("NOT POSSIBLE")
        else:
            if self.isEmpty():
                self.front=self.rear=0
                self.q[self.rear]=data
            else:
                self.rear=self.rear+1
                self.q[self.rear]=data

    def enqueue_front(self,data):
        if self.isFull():
            print("NOT POSSIBLE")
        else:
            if self.front==0:
                print("NOT POSSIBLE")
            else:
                self.front=self.front-1
                self.q[self.front]=data

    def dequeue(self):
        if self.isEmpty():
            print("NOT POSSIBLE")
        else:
            if self.front==self.rear:
                temp=self.q[self.front]
                self.front=self.rear=-1
                return temp
            else:
                temp=self.q[self.front]
                self.front=self.front+1
                return temp

    def dequeue_rear(self):
        if self.isEmpty():
            print("not possible")
        else:
            if self.front==self.rear:
                temp=self.q[self.rear]
                self.front=self.rear=-1
                return temp
            else:
                temp= self.q[self.rear]
                self.rear=self.rear-1
                return temp
    def display(self):
      print(self.q[self.front:self.rear+1])

q=doublequeue()
q.enqueue(1)
q.display()
q.enqueue(2)
q.enqueue(2)
q.enqueue(2)
q.enqueue(2)
q.display()
```

```python
#circlular queue
class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = self.rear = -1

    def is_empty(self):
        return self.front == -1

    def is_full(self):
        return (self.rear + 1) % self.size == self.front

    def enqueue(self, item):
        if self.is_full():
            print("Queue is full")
        elif self.is_empty():
            self.front = self.rear = 0
            self.queue[self.rear] = item
        else:
            self.rear = (self.rear + 1) % self.size
            self.queue[self.rear] = item

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
        elif self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.size

    def display(self):
        if self.is_empty():
            print("Queue is empty")
        elif self.front <= self.rear:
            for i in range(self.front, self.rear + 1):
                print(self.queue[i], end=" ")
            print()
        else:
            for i in range(self.front, self.size):
                print(self.queue[i], end=" ")
            for i in range(0, self.rear + 1):
                print(self.queue[i], end=" ")
            print()

# Example usage:
cq = CircularQueue(5)
cq.enqueue(1)
cq.enqueue(2)
cq.enqueue(3)
cq.enqueue(4)
cq.enqueue(5)

cq.display()  # Output: 1 2 3 4 5

cq.dequeue()
cq.dequeue()

cq.display()  # Output: 3 4 5
```

```python
#stack concept using linked list

class Node:
  def __init__ (self,data):
    self.data=data
    self.next=None
class stackll:
  def __init__(self):
    self.head=None
  def push(self,data):
    newnode=Node(data)
    if self.head==None:
      self.head=newnode
    else:
      newnode.next=self.head
      self.head=newnode
  def pop(self):
    if self.head==None:
      print("NOT POSSIBLE")
    else:
      temp=self.head.data
      self.head=self.head.next
      return temp
s=stackll()
for i in range(1,11):
  s.push(i)
for i in range(10):
  print(s.pop(),end=" ")
```
